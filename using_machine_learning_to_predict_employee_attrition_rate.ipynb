```python
ordinal_features = ['EducationField','Department','JobRole','MaritalStatus','Gender','Attrition']
data[ordinal_features] = data[ordinal_features].astype('category')
data[ordinal_features] = data[ordinal_features].apply(lambda x: x.cat.codes)
data.head()
```

```python
nominal_features = ['BusinessTravel']
data[nominal_features] = data[nominal_features].astype('category')
data[nominal_features] = data[nominal_features].apply(lambda x: x.cat.codes)
data.head()
```

```python
edu_map = {1:'Below College',2:'College',3:'Bachelor',4:'Master',5:'Doctor'}
plt.figure(figsize=(12,5))
sns.countplot(x='EducationField',hue='Attrition',data=data,palette='hot')
plt.title("Education w.r.t Attrition")
plt.xticks(rotation=45)
plt.show()
```

```python
data['Attrition'] = data['Attrition'].replace({'No':0,'Yes':1})
data.head()
```

```python
data['OverTime'] = data['OverTime'].replace({'No':0,'Yes':1})
data['Gender'] = data['Gender'].replace({'Male':0,'Female':1})
data.head()
```

```python
data['Over18'] = data['Over18'].replace({'N':0,'Y':1})
data.head()
```

```python
from sklearn.preprocessing import LabelEncoder
encoding_cols=['BusinessTravel','Department','EducationField','JobRole','MaritalStatus']
label_encoders = {}
for column in encoding_cols:
  label_encoders[column] = LabelEncoder()
  data[column] = label_encoders[column].fit_transform(data[column])
```

```python
data.head()
```

```python
data.info()
```

```python
x = data.drop(['Attrition','Over18'], axis=1)
y = data['Attrition'].values
```

```python
from collections import Counter
from imblearn.over_sampling import RandomOverSampler
print(Counter(y))
rus = RandomOverSampler(random_state = 42)
x_over, y_over = rus.fit_resample(x,y)
print(Counter(y_over))
```

```python
from sklearn.model_selection import train_test_split
x_train, x_test, y_train, y_test = train_test_split(x_over, y_over, test_size=0.2, random_state=42)
```

```python
print(x_train.shape)
print(x_test.shape)
print(y_train.shape)
print(y_test.shape)
```

```python
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import confusion_matrix, accuracy_score, roc_curve, roc_auc_score
```

```python
logreg = LogisticRegression()
logreg.fit(x_train, y_train)
```

```python
prediction=logreg.predict(x_test)
cnf_matrix = confusion_matrix(y_test, prediction)
print("Accuracy Score-", accuracy_score(y_test, prediction))
```

```python
fig = plt.figure(figsize = (15,6))
ax1 = fig.add_subplot(1,2,1)
ax1 = sns.heatmap(pd.DataFrame(cnf_matrix), annot =True, cmap ='Blues', fmt ='d')
bottom, top = ax1.get_ylim()
ax1.set_ylim(bottom + 0.5, top - 0.5)
plt.xlabel('Predicted')
plt.ylabel('Expected')

ax2 = fig.add_subplot(1,2,2)
y_pred_prob = logreg.predict_proba(x_test)[:,1]
fpr, tpr, thresholds = roc_curve(y_test, prediction)
auc = roc_auc_score(y_test, prediction)
ax2 = plt.plot(fpr, tpr, label = "data 1, auc="+str(auc))
plt.legend(loc=4)
plt.show()
```
