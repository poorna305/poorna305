```python
import pandas as pd
```

```python
# series method
series = pd.Series([1,2,3,4,5])
print(series)
```

```python
series = pd.Series([1,2,3,4,5], index=['a','b','c','d','e'])
print(series)
```

```python
print(series['a'])
```

```python
print(series['d'])
```

```python
# data frame
data = pd.DataFrame([1,2,3,4,5,6,7])
print(data)
```
